# Detecting outliers in numerical columns using the Interquartile Range (IQR) method.
Note: This is only calculated for the numerical columns in df_deliveries2

```python
numeric = df_deliveries2.select_dtypes(include=['number'])

# Dictionary to keep outliers
outliers_dict = {}

for col in numeric.columns:
    Q1 = numeric[col].quantile(0.25)
    Q3 = numeric[col].quantile(0.75)
    IQR = Q3 - Q1

    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR

    # Outliers y column
    outliers = numeric[(numeric[col] < lim_inf) | (numeric[col] > lim_sup)]

    outliers_dict[col] = outliers[col]

# Show results
for col, vals in outliers_dict.items():
    print(f"Outliers en {col}:")
    print(vals)
    print("-"*40)
```

### Outlier Removal
We will remove the outliers detected in the `tiempo_entrega_minutos` and `calificacion_cliente` columns using the Interquartile Range (IQR) method. The data will be filtered to retain only the values that fall within the defined IQR boundaries ($Q1 - 1.5 \times IQR$ and $Q3 + 1.5 \times IQR$).


```python
# Create a copy of the DataFrame to avoid modifying the original one directly
df_deliveries_no_outliers = df_deliveries2.copy()

# Columns where outliers were identified and need to be removed
columns_outliers = ['tiempo_entrega_minutos']

for col in columns_outliers:
    # Calculate Q1, Q3 e IQR
    Q1 = df_deliveries_no_outliers[col].quantile(0.25)
    Q3 = df_deliveries_no_outliers[col].quantile(0.75)
    IQR = Q3 - Q1

   # Calculate operational boundaries for outlier detection
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR

# Outlier removal: Retain only values within the acceptable boundaries
    df_deliveries_no_outliers = df_deliveries_no_outliers[(df_deliveries_no_outliers[col] >= lim_inf) & (df_deliveries_no_outliers[col] <= lim_sup)]

print(f"DataFrame original: {len(df_deliveries2)} filas")
print(f"DataFrame without outliers: {len(df_deliveries_no_outliers)} rows")

# Display descriptive statistics for the columns after cleaning
print("\nDescriptive statistics for columns with outliers removed:")
print(df_deliveries_no_outliers[columns_outliers].describe())
```

# 3) Normalize/Standardize numerical variables
# 4) Split data into training (80%) and testing (20%) sets
# Result: Implemented a Logistic Regression model with a 96% accuracy

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

# Select only numerical columns for standardization.
# Exclude 'id_entrega', 'id_mensajero', 'fecha', 'zona', 'estado', 'calificacion_cliente', 'fecha_solo_fecha', 'dia_semana', 'hora', and the generated dummy columns.
# The target variable 'es_entrega_lenta' must also be temporarily excluded from X.

# Identify numerical columns that are not IDs, dates, or the target variable
numerical_columns = df_deliveries_encoded.select_dtypes(include=['float64', 'int64']).columns.tolist()

# Exclude the target variable and other non-relevant columns from X (if present in the numerical list)
# 'tiempo_entrega_minutos' is the basis for the target variable, so it should not be included as a feature
columns_to_exclude_from_features = ['es_entrega_lenta', 'tiempo_entrega_minutos']
numerical_features = [col for col in numerical_columns if col not in columns_to_exclude_from_features]

X_numeric = df_deliveries_encoded[numerical_features]

# Impute missing values in X_numeric BEFORE scaling
# Only calificacion_cliente has NaNs as per earlier analysis
if 'calificacion_cliente' in X_numeric.columns:
    imputer = SimpleImputer(strategy='mean')
    X_numeric['calificacion_cliente'] = imputer.fit_transform(X_numeric[['calificacion_cliente']])

# Standardize the numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)
X_scaled_df = pd.DataFrame(X_scaled, columns=numerical_features, index=X_numeric.index)

# Combine standardized features with the categorical dummy columns
# First, fetch the dummy columns. Ensure original columns that were dummy-encoded are excluded
dummy_cols = [col for col in df_deliveries_encoded.columns if col.startswith(('ciudad_', 'tipo_vehiculo_', 'dia_semana_'))]
X_final = pd.concat([X_scaled_df, df_deliveries_encoded[dummy_cols]], axis=1)

y = df_deliveries_encoded['es_entrega_lenta']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_final, y, test_size=0.2, random_state=42, stratify=y)

print("X_train dimensions:", X_train.shape)
print("X_test dimensions:", X_test.shape)
print("y_train dimensions:", y_train.shape)
print("y_test dimensions:", y_test.shape)

print("\nFirst 5 rows of X_train (standardized and encoded features):")
display(X_train.head())
```

3) Normaliza/estandariza las variables numéricas.
4)Divide los datos en entrenamiento (80%) y prueba (20%)R/ Se Realiza un modelo de regresion logistica con una precision del 96%

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import KFold, cross_val_score

models = {
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "Random Forest": RandomForestClassifier()
}

results = []
for model in models.values():
    kf = KFold(n_splits=6, random_state=42, shuffle=True)
    cv_results = cross_val_score(model, X_train, y_train, cv=kf, scoring="accuracy")
    results.append(cv_results)

plt.boxplot(results, tick_labels=models.keys())
plt.title("Model Cross-Validation Accuracy") # Added a title for clarity
plt.ylabel("Accuracy") # Added y-label
plt.xticks(rotation=45, ha='right') # Rotate labels for better readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

# Define the parameter distribution for RandomizedSearchCV
param_dist = {
    'n_estimators': randint(50, 200),
    'max_depth': randint(3, 20),
    'min_samples_leaf': randint(1, 10),
    'min_samples_split': randint(2, 20),
    'criterion': ['gini', 'entropy']
}

# Instantiate KFold
kf = KFold(n_splits=5, random_state=42, shuffle=True)

# Initialize a RandomForestClassifier model
rf_model = RandomForestClassifier(random_state=42)

# Instantiate RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=rf_model,
    param_distributions=param_dist,
    n_iter=50,  # Number of parameter settings that are sampled. Trade-off between runtime and quality of the solution.
    cv=kf,
    scoring='accuracy',
    random_state=42,
    n_jobs=-1  # Use all available cores
)

# Fit RandomizedSearchCV to the scaled training data
random_search.fit(X_train, y_train)

print("Best parameters:", random_search.best_params_)
print("Best accuracy:", random_search.best_score_)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Get the best parameters from RandomizedSearchCV
best_params = random_search.best_params_

# Initialize a RandomForestClassifier with the best parameters
modelo_rf = RandomForestClassifier(random_state=42, **best_params)

# Train the model
modelo_rf.fit(X_train, y_train)

# Make predictions on the scaled test data
y_pred_rf = modelo_rf.predict(X_test)

# Print the evaluation metrics
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nReporte de clasificación:")
print(classification_report(y_test, y_pred_rf))
print("\nMatriz de confusión:")
print(confusion_matrix(y_test, y_pred_rf))
```

```python
import joblib

# Save the trained Random Forest model
joblib.dump(modelo_rf, 'random_forest_model.joblib')
print("Modelo Random Forest guardado como 'random_forest_model.joblib'")

# Save the StandardScaler
joblib.dump(scaler, 'scaler.joblib')
print("Scaler guardado como 'scaler.joblib'")
```
